```python
import pandas as pd
import plotly.express as px

# 1. Load Data
df = pd.read_csv("hop_enrichment.csv")

# 2. Aggregate Unique IPs per Country
country_counts = df.dropna(subset=['country']).groupby('country')['ip_addr'].nunique().reset_index()
country_counts.columns = ['iso2', 'unique_ips']

# 3. Map ISO2 to ISO3
iso2_to_iso3 = {
    'NL': 'NLD', 'DE': 'DEU', 'AT': 'AUT', 'SE': 'SWE', 'US': 'USA', 'BE': 'BEL', 
    'FR': 'FRA', 'RU': 'RUS', 'AU': 'AUS', 'SG': 'SGP', 'CZ': 'CZE', 'GB': 'GBR', 
    'CH': 'CHE', 'ZA': 'ZAF', 'CA': 'CAN', 'IR': 'IRN', 'BG': 'BGR', 'LB': 'LBN', 
    'SK': 'SVK', 'IT': 'ITA', 'BR': 'BRA', 'MD': 'MDA', 'RO': 'ROU', 'BY': 'BLR', 
    'IN': 'IND', 'NO': 'NOR', 'PL': 'POL', 'IE': 'IRL', 'ZM': 'ZMB', 'LU': 'LUX', 
    'TN': 'TUN', 'VE': 'VEN', 'PR': 'PRI', 'JP': 'JPN', 'TW': 'TWN', 'FI': 'FIN', 
    'AZ': 'AZE', 'HU': 'HUN', 'DK': 'DNK', 'LV': 'LVA', 'KE': 'KEN', 'MX': 'MEX', 
    'ZW': 'ZWE', 'ES': 'ESP', 'PT': 'PRT', 'KZ': 'KAZ'
}
country_counts['country'] = country_counts['iso2'].map(iso2_to_iso3)
country_counts = country_counts.dropna(subset=['country'])

# 4. Create Interactive Map
fig = px.choropleth(
    country_counts,
    locations="country", 
    color="unique_ips",
    hover_name="iso2",
    projection="natural earth",
    # Bold title using HTML tag
    title="<b>Unique IP Addresses per Country</b>",
    color_continuous_scale=px.colors.sequential.Plasma,
    labels={'unique_ips': 'Unique IPs'}
)

# 5. Cosmetic Improvements
fig.update_layout(
    title_x=0.5, # Center the title
    margin={"r":0,"t":50,"l":0,"b":0}, # Minimize margins for immersive feel
    geo=dict(
        showframe=False,       # Remove the rectangular frame around the map
        showcoastlines=False,  # Remove coastlines for a cleaner look
        projection_type='natural earth'
    )
)

# Remove country borders
fig.update_traces(marker_line_width=0)

# 6. Save
fig.write_html("world_map.html")
```
