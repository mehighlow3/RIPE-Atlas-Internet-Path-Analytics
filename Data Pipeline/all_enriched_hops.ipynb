```python
import json
import csv

input_file = "enriched_traceroutes.jsonl"   # JSONL file
output_file = "all_hops.csv"   # resulting CSV

with open(input_file, "r", encoding="utf-8") as fin, \
     open(output_file, "w", newline="", encoding="utf-8") as fout:

    writer = csv.writer(fout)
    writer.writerow(["msm_id", "prb_id", "hop", "ip_addr", "asn", "country"])

    for line in fin:
        doc = json.loads(line)

        msm_id = doc.get("msm_id")
        prb_id = doc.get("prb_id")

        for hop_entry in doc.get("result", []):
            hop = hop_entry.get("hop")
            results = hop_entry.get("result", [])

            # Keep ONLY the second result (index 1)
            if len(results) >= 2:
                r = results[1]

                writer.writerow([
                    msm_id,
                    prb_id,
                    hop,
                    r.get("ip_addr"),
                    r.get("asn"),
                    r.get("country")
                ])

print("Making of CSV that has all hops (including repeted ones) finished.")
```
